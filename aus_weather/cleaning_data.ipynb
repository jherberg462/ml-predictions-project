```python
#start feature engineering
df = None
df = pd.read_csv('weatherAUS.csv')
```

```python
#clean up data

#create column with True value if RainToday == 'Yes', and false otherwise
df['rain_today_b'] = df['RainToday'].apply(lambda x: True if x == "Yes" else False)
#same as above, except via the RainTomorrow column
df['rain_tomorrow_b'] = df['RainTomorrow'].apply(lambda x: True if x == "Yes" else False)
#drop RainTomorrow and RainToday columns
df = df.drop(columns=['RainTomorrow', 'RainToday'])
#drop columns that should have no affect on outcome
df = df.drop(columns=['Date', 'Location'])
#drop column that leaks information to model
df = df.drop(columns=['RISK_MM'])
```

```python
#start filling na values

#
df['Rainfall'] = np.where(df['Rainfall'] == np.NaN, np.where(df['rain_today_b'] == False, 0, np.nan), df['Rainfall'])
```

```python
#fill na values in the evaporation column with the mean values of this column 
df['Evaporation'] = df['Evaporation'].fillna(df['Evaporation'].mean())
```

```python
df['Cloud9am'] = df['Cloud9am'].fillna(df['Cloud9am'].mean())
df['Cloud3pm'] = df['Cloud3pm'].fillna(df['Cloud3pm'].mean())
```

```python
#get number of null values in the Sunshine column
df['Sunshine'].isnull().value_counts()
#not sure we want to not drop the na values in this column due to large importance in models
# df['Sunshine'] = df['Sunshine'].fillna(df['Sunshine'].mean())
#removing na values reduces the accuracy of the models
#the null values in this column make up the majority of rows we are dropping
#will not get much more data by filling additional na values
```

```python
# drop rows in which at least one value is null
df = df.dropna()
df.count()
```

```python
#create column with difference between tempeture measurements
#changing to absolute value increases model performance
df['temp_change_9to3'] = abs(df['Temp9am'] - df['Temp3pm'])
```

```python
#changing from difference to absolute difference increases performance
df['temp_change_min_max'] = abs(df['MaxTemp'] - df['MinTemp'])
```

```python
#changing to absolute difference decreases model accuracy (decreases even more when incorporated into percent column below)
df['humidity_change'] = df['Humidity9am'] - df['Humidity3pm']
```

```python
df['humidity_change_percent'] = np.where(df['Humidity9am'] > 0,df['humidity_change'] / df['Humidity9am'], False )
df['humidity_change_percent'] = np.where(df['humidity_change_percent'] == False, df['humidity_change_percent'].max(), df['humidity_change_percent'])
```

```python
#changing to absolute difference decreases model accuracy
df['pressure_change'] = df['Pressure9am'] - df['Pressure3pm']
```

```python
df['wind_change_direction'] = np.where(df['WindDir9am'] == df['WindDir3pm'], True, False)
```

```python
df['wind_gust_change_3'] = np.where(df['WindGustDir'] == df['WindDir3pm'], False, True)
df['wind_gust_change_9'] = np.where(df['WindGustDir'] == df['WindDir9am'], False, True)
```

```python
df = df.drop(columns=['WindDir9am', 'WindDir3pm', 'WindGustDir'])
```
