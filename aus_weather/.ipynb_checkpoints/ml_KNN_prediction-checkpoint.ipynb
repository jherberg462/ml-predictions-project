```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import os
```

```python
df = pd.read_csv(os.path.join("weatherAUS_clean.csv"))
df.head(10)
```

```python
list(df.columns)
```

```python
# Drop RISK_MM column from the dataset
df1 = df.drop("RISK_MM", axis=1)
df1.head()
```

```python
y = df1["rain_tomorrow_b"]

target_names = ["yes", "no"]
```

```python
X = df1.drop("rain_tomorrow_b", axis=1)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 40, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_score = knn.score(X_train, y_train)
    test_score = knn.score(X_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 40, 2), train_scores, marker='o')
plt.plot(range(1, 40, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score")
plt.show()
```

```python
# Choose the best k from above and re-fit the KNN Classifier using that k value.
# print the score for the test data
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train, y_train)
print('k=30 Test Acc: %.3f' % knn.score(X_test, y_test))
```

## Improving the model

```python
# import joblib
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=200)
model.fit(X_train, y_train)
```

```python
#get value of each feature in a list
feature_value = list(model.feature_importances_)
#list of feature names
feature_names = list(df1.columns)
#loop to zip the feature name and its value together
features_and_values = [(feature, importance) for feature, importance in zip(feature_names, feature_value)]
features_and_values
sorted_list = sorted(features_and_values, key = lambda x: x[1], reverse = False)
```

```python
# Getting list of columns to be dropped
columns_list = []

values_sum = 0

for i in sorted_list:
    values_sum += i[1]
    columns_list.append(i[0])
    if values_sum >= 0.05:
        break
print(columns_list)

print(len(sorted_list))
print(len(columns_list))
```

```python
data_shorted = df1.drop(columns_list, axis =1)
data_shorted.head()
```

```python
X1 = data_shorted
X1.head()
```

```python
# X1 with the new data_shorted
from sklearn.model_selection import train_test_split
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, random_state=42)
```

```python
# Loop through different k values to see which has the highest accuracy
# Note: We only use odd numbers because we don't want any ties
train_scores = []
test_scores = []
for k in range(1, 40, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X1_train, y_train)
    train_score = knn.score(X1_train, y_train)
    test_score = knn.score(X1_test, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f"k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}")
    
    
plt.plot(range(1, 40, 2), train_scores, marker='o')
plt.plot(range(1, 40, 2), test_scores, marker="x")
plt.xlabel("k neighbors")
plt.ylabel("Testing accuracy Score(data_shorted)")
plt.show()
```

```python
# Choose the best k from above and re-fit the KNN Classifier using that k value.
# print the score for the test data
knn = KNeighborsClassifier(n_neighbors=30)
knn.fit(X_train, y_train)
print('k=30 Test Acc: %.3f' % knn.score(X_test, y_test))
```
